```python
year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
month = ['01','02','03','04','05','06','07','08','09','10','11','12']

date = []
for y in year:
    for m in month:
        d = str(y)+m
        date.append(d)

price = []
import random
for _ in range(len(date)):
    p = random.randrange(1,100)
    price.append(p)

file = open("./csv01.csv", 'w')

for i in range(len(date)):
    msg = str(date[i]) + "," + str(price[i]) + "\n"
    file.write(msg)

file.close()
```

```python
import csv

f = open('./test01.csv')

No = []
Location = []
Name = []
Address = []
Date = []
Brand = []
Type = []
PremiumGasoline = []
Gasoline = []
Diesel = []
Kerosene = []

csvReader = csv.reader(f)

for row in csvReader:
    
    try:
        No.append(row[0])
        Location.append(row[1])
        Name.append(row[2])
        Address.append(row[3])
        Date.append(row[4])
        Brand.append(row[5])
        Type.append(row[6])
        PremiumGasoline.append(row[7])
        Gasoline.append(row[8])
        Diesel.append(row[9])
        Kerosene.append(row[10])
    except:
        pass
f.close()
```
