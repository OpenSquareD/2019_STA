# 서울시 기간별 시간평균 대기환경 정보 데이터 분석 및 시각화

## 데이터 다운로드
> https://data.seoul.go.kr/dataList/datasetView.do?infId=OA-2221&srvType=S&serviceKind=1&currentPageNo=1#AXexec


```python
import pandas as pd
```

```python
df = pd.read_csv('./서울시 기간별 시간평균 대기환경 정보.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df['측정일시']= pd.to_datetime(df['측정일시'],format='%Y%m%d%H%M')
```

```python
df.info()
```

```python
df['측정소명'].unique(), len(df['측정소명'].unique())
```

```python
df.groupby(by='측정소명')['초미세먼지(㎍/㎥)'].mean()
```

```python
df.groupby(by=['측정일시','측정소명'])['초미세먼지(㎍/㎥)'].mean()
```

```python
df_microdust = df.pivot_table(index="측정일시", columns="측정소명", values="초미세먼지(㎍/㎥)")
df_dust = df.pivot_table(index="측정일시", columns="측정소명", values="미세먼지 1시간(㎍/㎥)")
```

```python
df_microdust.to_csv('./서울시 기간별 시간평균 대기환경 정보(초미세먼지).csv', encoding='euc-kr')
df_dust.to_csv('./서울시 기간별 시간평균 대기환경 정보(미세먼지).csv', encoding='euc-kr')
```

```python
import plotly.graph_objects as go

# 구별 Line Chart
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_dust.index, y=df_dust['강남구'],
                    mode='lines', name='강남구 미세먼지'))

fig.add_trace(go.Scatter(x=df_dust.index, y=df_dust['용산구'],
                    mode='lines+markers', name='용산구 미세먼지'))

fig.show()
```

```python
df_microdust_mean = df.groupby(by='측정소명')['초미세먼지(㎍/㎥)'].mean()
```

```python
fig = go.Figure([go.Bar(x=df_microdust_mean.index, y=df_microdust_mean)])
fig.show()
```

```python
df_co_mean = df.groupby(by='측정소명')['일산화탄소농도(ppm)'].mean()
```

```python
fig = go.Figure([go.Bar(x=df_co_mean.index, y=df_co_mean)])
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y=df_dust['강남구'], name='강남구'))
fig.add_trace(go.Box(y=df_dust['강서구'], name='강서구'))
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

for xd, yd in zip(df_dust.columns, df_dust):
        fig.add_trace(go.Box(y=df_dust[yd], name=xd))
        
fig.update_layout(
    title='서울특별시 구별 미세먼지 분포 (20190616~2019816)')

fig.show()
```
