```python
x = df.iloc[:, :-1].astype('str')
y = df.iloc[:,-1].str.replace(',','').astype('float')
```

```python
x_dummi = pd.get_dummies(x)
x_dummi.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_dummi, y, test_size = 0.3, random_state = 5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, Y_train)

lm.score(X_test, Y_test)
```

```python
Y_pred = lm.predict(X_test)
```

```python
gap = []
for real, pred in zip(Y_test, Y_pred):
    print('실제 값 :', real, '예측 값 : ', pred, '차이 :', real-pred)
    gap.append(real-pred)
```

```python
np.mean(gap)
```
