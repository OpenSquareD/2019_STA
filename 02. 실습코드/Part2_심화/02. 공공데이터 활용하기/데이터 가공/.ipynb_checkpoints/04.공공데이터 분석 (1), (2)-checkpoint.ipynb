# 공공데이터 분석 1

### 학습 목표
- Jupyter Notebook 환경에서 Pandas와 Matplotlib 라이브러리를 활용하여 공공 데이터를 분석
- 서울시 CCTV 설치 현황 데이터와 인구 통계 데이터를 분석

### 사용 데이터
- 서울시 자치구 연도별 cctv 설치 현황
- 서울시 인구 통계

공공데이터 파일(CSV)을 읽기 위해 Pandas 라이브러리를 불러옴.

```python
import pandas as pd
```

Pandas 라이브러리에서 read_csv 함수를 이용하여 csv 파일을 읽을 수 있음.

또한, head 함수를 통해 csv 파일이 제대로 읽혔는지 확인.

```python
cctv = pd.read_csv('./data/seoulcctv.csv',  encoding='utf-8')
cctv.head()
```

Columns 속성을 통해 불러온 CSV 파일의 데이터 확인

```python
cctv.columns
```

기관명 코드를 보다 적합하게 지역구로 컬럼의 이름을 변경

또한, head 함수를 통해 변경 확인

```python
cctv.columns=[ '지역구', '소계',  '2013년도 이전', '2014년', '2015년', '2016년']
cctv.head()
```

CCTV 설치 현황 데이터와 동일하게 서울시 인구 통계 정보 데이터도 읽어옴.

서울시 인구 통계 정보는 excel 파일이며 read_excel 함수를 이용함.
> read_excel 함수의 경우 xlrd 라이브러리가 필요하며, 해당 라이브러리가 설치가 안되어 있는 경우 pip 명령을 통해 라이브러리를 설치함

head 함수를 통해 해당 파일이 정상적으로 불러왔는지 확인

```python
people = pd.read_excel('./data/seoulpeople.xls',  encoding='utf-8')
people.head()
```

데이터가 정상적으로 불러왔으나 몇가지 문제점이 존재

- 열의 이름이 3행에 걸쳐 작성되어 있음
- B(자치구), D(합계), G(한국인), J(외국인), N(고령자) 열만 필요

따라서 read_excel 함수의 옵션을 이용하여 원하는 데이터 프레임 형식으로 파일을 읽음

```python
people = pd.read_excel('./data/seoulpeople.xls', 
                          header = 2,
                          usecols = 'B, D, G, J, N', #자치구,합계,한국인,외국인,고령자
                          encoding='utf-8')
people.head()
```

원하는 데이터 프레임 형식으로 파일이 읽어 왔으나 열의 이름을 알아보는데 문제가 발생

따라서, 열의 이름을 변경

```python
people.columns=[ '지역구', '인구수',  '한국인', '외국인', '고령자']
people.head()
```

데이터 프레임 함수인 sort_values 함수를 통해 CCTV 설치 하위 5개 구와 상위 5개 구를 확인

```python
cctv.sort_values(by='소계', ascending=True).head(5)
```

```python
cctv.sort_values(by='소계', ascending=False).head(5)
```

인구 비율에 따른 CCTV 설치 수를 계산하기 위해 인구 비율을 계산 하여 새로운 열을 생성

```python
people['외국인비율'] = people['외국인'] / people['인구수'] * 100
people['고령자비율'] = people['고령자'] / people['인구수'] * 100
people.head()
```

불필요한 행(합계)를 drop 함수를 통해 제거
inplace 옵션이 True인 경우 원본 데이터 프레임 내부에서 삭제
inplace 옵션이 False인 경우 해당 열을 제거한 새로운 데이터 프레임을 반환

```python
people.drop([0], inplace=True)
people.head()
```

외국인 비율의 상위 5개 구와 하위 5개 구를 확인

```python
people.sort_values(by='외국인비율', ascending=True).head(5)
```

```python
people.sort_values(by='외국인비율', ascending=False).head(5)
```

CCTV와 인구 통계 데이터 결합

```python
df = pd.merge(cctv, people, on='지역구')
df.head()
```

필요 없는 열을 삭제

```python
del df['2013년도 이전']
del df['2014년']
del df['2015년']
df.head()
```

이후, 재구현성을 위해 가공이 완료된 데이터 프레임을 저장함.

```python
df.to_csv("./cctv_people.csv")
```

보다 효과적인 설득 효과를 위해 분석한 결과를 시각화

시각화를 위해 matplotlib 라이브러리를 불러옴

```python
import matplotlib.pyplot as plt
```

지역구별 CCTV 설치수와 인구수, 외국인 비율, 고령자 비율의 상관 관계 계산

상관 관계 계산은 Numpy 라이브러리에 있는 corrcoef 함수를 통해 계산

```python
import numpy as np
np.corrcoef(df['고령자비율'],df['소계'])
```

```python
np.corrcoef(df['외국인비율'],df['소계'])
```

```python
np.corrcoef(df['인구수'],df['소계'])
```

CCTV 설치수와 고령자 비율 및 외국인 비율은 음의 상관 관계가 있으며
CCTV 설치수와 인구수는 양의 상관 관계가 있음

하지만, 상관계수의 값이 작기 때문에 영향도가 있다고 보기는 힘듬 

```python
plt.figure()
df['소계'].plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

Y축의 Index 확인의 어려움이 존재 

따라서 데이터 프레임의 Index를 지역구로 변경

```python
df.set_index('지역구', inplace=True)
df.head()
```

이후, Matplotlib로 재 시각화
Matplotlib에서 한글을 사용할 경우 에러가 발생 할 수 있음.
따라서, Font를 설정

```python
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
df['소계'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

인구수와 CCTV 설치 수를 통해 인구수 대비 CCTV 비율을 계산 및 시각화

```python
df['CCTV비율'] = df['소계'] / df['인구수'] * 100
df['CCTV비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

# 공공데이터 분석 2

### 학습 목표

서울시 범죄 통계 데이터와 인구 통계 데이터를 분석

### 사용 데이터

서울시 범죄 통계 데이터

서울시 인구 통계 데이터

서울열린데이터광장에서 제공하는 서울시 5대 범죄 발행 현황 통계데이터를 다운로드

다운로드한 파일을 Pandas 라이브러리를 이용하여 불러옴

```python
crime = pd.read_excel('./data/crime.xls',  encoding='utf-8')
crime.head()
```

인구 통계 데이터와 동일하게 불필요한 열을 제외하고 원하는 데이터만 불러옴

또한 drop 함수를 이용하여 불필요한 행 제거 및 열 이름 변경

```python
crime = pd.read_excel('./data/crime.xls',  
                      usecols = 'B, E, F, G, H, I, J, K, L, M, N', 
                      encoding='utf-8')
crime.head()
```

불필요한 행 제거 및 열 이름 재설정

```python
crime.drop([0], inplace=True)
crime.drop([1], inplace=True)
crime.columns=[ '지역구', '살인 발생', '살인 검거', '강도 발생', '강도 검거', '강간강제추행 발생', '강간강제추행 검거','절도 발생','절도 검거','폭력 발생','폭력 검거']

crime.head()
```

데이터 프레임 기초 통계 분석

```python
crime.describe()
```

특정 문자('-') 치환

```python
crime = crime.replace('-',int(0))
```

검거 횟수와 발생 횟수를 이용하여 검거율 계산

```python
crime['강간강제추행 검거율'] = crime['강간강제추행 검거']/crime['강간강제추행 발생']*100
crime['강도 검거율'] = crime['강도 검거']/crime['강도 발생']*100
crime['살인 검거율'] = crime['살인 검거']/crime['살인 발생']*100
crime['절도 검거율'] = crime['절도 검거']/crime['절도 발생']*100
crime['폭력 검거율'] = crime['폭력 검거']/crime['폭력 발생']*100

crime.head()
```

검거율이 100%를 초과하는 경우 100으로 치환

```python
con_list = ['강간강제추행 검거율', '강도 검거율', '살인 검거율', '절도 검거율', '폭력 검거율']

for column in con_list:
    crime.loc[crime[column] > 100, column] = 100
    
crime.head()
```

불필요한 열 제거

```python
del crime['강간강제추행 검거']
del crime['강도 검거']
del crime['살인 검거']
del crime['절도 검거']
del crime['폭력 검거']
crime.head()
```

열 이름 재 정리 및 Index 재설정 

```python
crime.rename(columns = {'강간강제추행 발생':'강간',
                        '강도 발생':'강도', 
                        '살인 발생':'살인', 
                        '절도 발생':'절도', 
                        '폭력 발생':'폭력'}, inplace=True)

crime.set_index('지역구', inplace=True)
crime.head()

crime.head()
```

Scikit Learn 라이브러리에 존재하는 정규화 함수를 이용하여 정규화 진행
> 정규화 : 0~1사이의 값으로 변경

```python
from sklearn import preprocessing

col = ['강간', '강도', '살인', '절도', '폭력']

x = crime[col].values
min_max_scaler = preprocessing.MinMaxScaler()

x_scaled = min_max_scaler.fit_transform(x.astype(float))
crime_norm = pd.DataFrame(x_scaled, columns = col, index = crime.index)

col2 = ['강간강제추행 검거율', '강도 검거율', '살인 검거율', '절도 검거율', '폭력 검거율']
crime_norm[col2] = crime[col2]
crime_norm.head()
```

범죄 열 생성

```python
col = ['강간','강도','살인','절도','폭력']
crime_norm['범죄'] = np.sum(crime_norm[col], axis=1)
crime_norm.head()
```

검거 열 생성

```python
col = ['강간강제추행 검거율','강도 검거율','살인 검거율','절도 검거율','폭력 검거율']
crime_norm['검거'] = np.sum(crime_norm[col], axis=1)
crime_norm.head()
```

시각화를 위해 Matplotlib 라이브러리와 Seaborn 라이브러리를 불러옴

한글 사용을 위해 폰트 설정

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import platform

path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

Seaborn 라이브러리의 Pairplot 을 이용하여 강도, 살인, 폭력 시각화 

```python
sns.pairplot(crime_norm, vars=["강도", "살인", "폭력"], kind='reg', size=3)
plt.show()
```

검거 열 정규화

```python
tmp_max = crime_norm['검거'].max()
crime_norm['검거'] = crime_norm['검거'] / tmp_max * 100
crime_norm = crime_norm.sort_values(by='검거', ascending=False)
crime_norm.head()
```

검거율 정보 시각화(Heatmap)

```python
target_col = ['강간강제추행 검거율', '강도 검거율', '살인 검거율', '절도 검거율', '폭력 검거율']

crime_norm = crime_norm.sort_values(by='검거', ascending=False)

plt.figure(figsize = (10,10))
sns.heatmap(crime_norm[target_col], annot=True, fmt='f', 
                    linewidths=.5, cmap='RdPu')
plt.title('범죄 검거 비율 (정규화된 검거의 합으로 정렬)')
plt.show()
```
